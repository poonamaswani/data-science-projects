Customer segmentation allows a business to group customers based on demographics (e.g. age, gender, education, occupation, marital status, family size), geographics (e.g. country, transportation, preferred language), psychographics (e.g. lifestyle, values, personality, attitudes), behaviour (e.g. purchase history, brand loyalty, response to marketing activities), technographic (e.g. device type, browser type, original source), and needs (e.g. product features, service needs, delivery method). Since we have a transnational data set, which implies customers from different continents, several metrics are important when performing customer segmentation for target marketing. From a marketing perspective, the following five metrics help to understand the nuance of the customer base, buying behaviour, preferences, and value to the business.
- **Frequency** indicates how often a customer purchases over a given period of time. A high frequency indicates a loyal customer, a high level of satisfaction, trust or brand loyalty, and/or effective marketing efforts. Frequency based on purchases guides a business in the effectiveness of target marketing campaigns and how to target less active customers.
- **Recency** measures how recently a customer made a purchase or placed an order. It helps predict customer churn (turnover) and engagement. - **Customer lifetime value (CLV)** indicates the average or total value a customer contributes to a business over the course of their relationship. In other words, CLV is a metric of the total income a business can expect to generate from a customer as long as said customer remains a loyal client. CLV helps to prioritise marketing efforts and resources as it focuses on customers who are expected to bring the most value over time. Therefore, retaining high-value customers.
- The **average unit cost** indicates if the customer prefers low cost or high cost items. This is related to the profitability of purchases. Customers buying products with a higher average unit cost price should be targeted differently. Customer segmentation assists in identifying these customers.

### Initial data exploration and assessment

```python
#Step 1: Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from rich.console import Console

console = Console()

#Step 1: Import the data with a url link from GitHub
data = pd.read_csv('https://github.com/FourthRevGitHub/CAM_DS_Public-Data-Sets/raw/main/Course%201/Week%206/CUSTOMERS_CLEAN.zip')

#Step 2: View DataFrame
print(data.shape)
data.head()
```

```python
#Step 2: Check missing values and duplicate rows
print(data.isnull().sum())
print("Duplicate Rows: ", data.duplicated().sum())
```

```python
#Step 3: Perform Feature Engineering

#Create Frequency feature : indicates how often a customer purchases over a given period of time
data['Frequency'] = data.groupby( 'Customer ID' )['Order ID'].transform('count')

#Create Recency feature : measures how recently a customer made a purchase or placed an order
data['Recency'] = (pd.to_datetime('today') - pd.to_datetime(data['Delivery_Date'])).dt.days

# Create Customer-Age feature
data['DOB'] = pd.to_datetime(data[ 'Customer_BirthDate' ], format='%d%b%Y') # Convert the DOB to datetime
today = pd.Timestamp( 'now')
data['Age'] = (today - data[ 'DOB']).dt.days // 365

# Create Average Unit Cost: indicates the profitability of purchases
data['Avg_Unit_Cost'] = pd.to_numeric(data['Unit Cost'].astype(str).str.replace('[$,]', '', regex=True))

# Create Customer Lifetime Value : indicates the average or total value a customer contributes to a business over the course of their relationship
data['CLV' ] = pd.to_numeric(data['Total Revenue'].astype(str).str.replace('[$,]', '', regex=True))

df_aggr = data.drop(columns=['City','Continent','Postal_Code','State_Province','Order_Date','Delivery_Date','Total Revenue',
                      'Unit Cost', 'Order ID','Profit','Days to Delivery','Loyalty Num','Customer_BirthDate',
                      'OrderTypeLabel','CustomerCountryLabel','Customer_Group','Customer_Type','DOB','Discount'])
```

```python
#Create an aggregated dataset

df_aggr_1 = df_aggr.groupby('Customer ID').agg({
    'Age': 'max',
    'Avg_Unit_Cost': 'mean',
    'CLV': 'sum',
    'Frequency': 'mean',
    'Recency': 'min'
}).reset_index()

print(df_aggr_1.shape)
df_aggr_1.info()
df_aggr_1.head()
```

```python
#Exploratory Data Analysis

df_working_set = df_aggr_1.drop(columns=['Customer ID'])

#Step 2: Histogram (distribution of features)
df_working_set.hist(figsize=(10,10), bins=50, edgecolor='black')
plt.title("Feature Distribution")
plt.show()

#Step 2: Check missing values and duplicate rows
print(df_working_set.isnull().sum())
print("Duplicate Rows: ", df_working_set.duplicated().sum())

#Step 3 : Generate descriptive statistics of data: Mean//Median/Identify 95th percentile of atleast two features
df_working_set.describe(percentiles=[.25, .50, .75, .95]).round(2)
```

```python
#BoxPlot to spot outliers
for col in df_working_set.columns:
  plt.figure(figsize=(6,4))
  sns.boxplot(y=df_working_set[col])
  plt.title(f"BoxPlot of {col}")
  plt.show()
```

```python
#Find outliers using One-Class SVM
from sklearn  import svm
from sklearn.preprocessing import StandardScaler

#Scale data for better SVM Performance
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_working_set)

#Use one class SVM method
model = svm.OneClassSVM(nu=0.01, kernel="rbf", gamma=0.5)
model.fit(X_scaled)

#Predict the class (normal or anomaly)
y_pred = model.predict(X_scaled)
print(y_pred)

#Create Dataframe
df_svm_model = pd.DataFrame(X_scaled, columns=df_working_set.columns)
df_svm_model["y_pred"] = y_pred
df_svm_model.head()

#Return the outliers
anomalies = df_svm_model[df_svm_model["y_pred"] == -1]
print("Anomalies Count: ",anomalies.shape[0])
```

*   **21** Duplicate rows found with the initial data set
*   After adding **5** new features and doing aggregate on Customer ID, we removed the columns that are of not use. The working data set contains **68300** rows (one row per customer ID)
*   No duplicate rows found after aggregating the 5 features on Customer ID

###Clustering with ML models

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

X = X_scaled
inertia=[]

for k in range(1,20):
  kmeans = KMeans(n_clusters=k, random_state=42)
  kmeans.fit(X)
  inertia.append(kmeans.inertia_)

plt.plot(range(1,20), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
range_n_clusters = [4,5,6,7]

for n_clusters in range_n_clusters:
    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10, n_init = 10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

X_small_df = df_working_set.sample(n=10000).reset_index(drop=True)

# Hierarchical clustering from scratch
def hierarchical_clustering(X_small_df, method='average'):
    dists = pdist(X_small_df, metric='euclidean')
    linkage_matrix = linkage(dists, method=method)
    return linkage_matrix

# Compute the linkage matrix
linkage_matrix = hierarchical_clustering(X_small_df)

# Number of clusters we want to extract
k = 5

# Retrieve the clusters
clusters = fcluster(linkage_matrix, k, criterion='maxclust')

# Plot the dendrogram
plt.figure(figsize=(15, 10))
dendrogram(linkage_matrix)
plt.title('Dendrogram')
plt.xlabel('Sample index or (Cluster size)')
plt.ylabel('Distance')
plt.show()
```

**Observations after performing Clustering with ML Models**


*   Elbow method performed using range **1-10**, **1-15** and **1-20**. No of clusters determined from plot has been **5** or **6** in all 3 different parameter scenarios
*   Silhouette Score was calculated for 4,5,6,7 number of clusters and the resultant score has been pretty much steady around **0.24**
*    The Silhouette score is pretty similiar for differernt number of clusters, which means there is **no clear separation** and the data is continous
*   Hierarchial clustering has been performed using "**5**" as optimal number of clusters - as after using different range of clusters in elbow method we saw 5 as the standard and also silhouette score remains constant around the range 4 to 7 number of clusters.
*   Since the notebook has been crashing while creating dendrogram, we selected a small working set of **10000** rows to do hierarchial clustering and dendrogram.
*   The dendrogram revealed distinct hierarchial relationships between customer groups.

### Dimensionality Reduction

```python
#scale the X_small_df data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_small_df_scaled = scaler.fit_transform(X_small_df)

#Apply PCA for visualisation
pca = PCA(n_components = 2)
X_small_df_pca = pca.fit_transform(x_small_df_scaled)
print(X_small_df_pca)

# Apply K-Means clustering to the scaled sample data
from sklearn.cluster import KMeans
# Assuming k=5 based on earlier analysis, you can change this if needed
kmeans_sample = KMeans(n_clusters=5, random_state=42, n_init=10)
sample_cluster_labels = kmeans_sample.fit_predict(x_small_df_scaled)

plot_df = pd.DataFrame(X_small_df_pca, columns=["PC1", "PC2"])
# Assign the cluster labels directly to the plot_df
plot_df["clusters"] = sample_cluster_labels
plot_df.head()

#Visualise the scatter plot for plot_df dataset
plt.figure(figsize=(8,6))
# Use the cluster labels for coloring, mapping them to a color palette
plt.scatter(plot_df["PC1"], plot_df["PC2"], c=plot_df["clusters"], cmap='viridis', alpha=0.7)
plt.title("PCA Results")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.colorbar(label='Cluster') # Add a colorbar to show the mapping of colors to clusters
plt.show()
```

```python
#use t-SNE dimensional reduction on X_small_df
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=42)
X_small_df_tsne = tsne.fit_transform(x_small_df_scaled)

plot_df_tsne = pd.DataFrame(X_small_df_tsne, columns=["TSNE1", "TSNE2"])
plot_df_tsne["clusters"] = sample_cluster_labels
plot_df_tsne.head()

plt.figure(figsize=(8,6))
plt.scatter(plot_df_tsne["TSNE1"], plot_df_tsne["TSNE2"], c=plot_df_tsne["clusters"], cmap='viridis', alpha=0.7)
plt.title("t-SNE Results")
plt.xlabel("TSNE1")
plt.ylabel("TSNE2")
plt.colorbar(label='Cluster') # Add a colorbar to show the mapping of colors to clusters
plt.show()
```

*   The **PCA** scatterplot shows clusters somewhat overlapping, which suggests that while PCA is good for dimensionality reduction, it does not separate the customer groups.
*   This overlap indicates that customers may share similarities across multiple features, and their segmentation boundaries are not strictly linear.
*   The spread of points along PC2 shows variation in customer behavior.
*   The **t-SNE** plot provides much clearer separation of clusters compared to PCA, highlighting that t-SNE captures the non-linear relationships better.
*   Clusters (especially **yellow and teal**) form more distinct, non-overlapping groups, which may indicate strong differences in their buying behavior or demographics.
*   The circular pattern in t-SNE suggests the presence of dense “core” groups.
